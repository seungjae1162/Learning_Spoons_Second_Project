```python
df = pd.read_csv('data/time.csv')
df
```

```python
df_region = df.loc[:,['date','Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan',
       'Sejong', 'Gyeonggi-do', 'Gangwon-do', 'Chungcheongbuk-do',
       'Chungcheongnam-do', 'Jeollabuk-do', 'Jeollanam-do', 'Gyeongsangbuk-do',
       'Gyeongsangnam-do', 'Jeju-do']]
df_region
```

```python
df_region.columns = ['date','서울특별시','부산광역시','대구광역시',
                 '인천광역시','광주광역시','대전광역시','울산광역시',
                 '세종특별자치시','경기도','강원도','충청북도','충청남도',
                 '전라북도','전라남도','경상북도','경상남도','제주특별자치도']
```

```python
data = pd.melt(df_region, id_vars=['date'])
data
data = data.sort_values(by='date')
```

```python
data.head()
```

```python
data.columns=['date','지역','확진자']
```

```python
px.bar(
    data,
    x = '지역',
    y = '확진자',
    animation_frame='date',
    color='지역'
)
```
