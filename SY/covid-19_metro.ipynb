```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font='Malgun Gothic')
```

```python
# 2019년도 12월
metro_1912 = pd.read_csv('./data/CARD_SUBWAY_MONTH_201912.csv', encoding='cp949')
metro_1912.tail()
```

```python
plt.figure(figsize=(30, 4))
plt.xticks(rotation=90)
sns.barplot(data=metro_1912, x='노선명', y='승차총승객수', ci=None)
```

```python
plt.figure(figsize=(30, 4))
plt.xticks(rotation=90)
sns.boxplot(data=metro_1912, x='노선명', y='승차총승객수')
```
