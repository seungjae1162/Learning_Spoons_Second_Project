```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
type(soup)
```

```python
# print(soup)
```

```python
sum1 = soup.select('tr.summary.first > td')

sum1_result = []

for td in sum1:
    sum1_txt = td.text
    sum1_result.append(sum1_txt)
    print(sum1_txt)
    if sum1_txt == '#U17':
        break
```

```python
def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))
```

```python
sum1_results = []

for group1 in chunker(sum1_result, 5):
    group1 = group1
    sum1_results.append(group1)
    print(group1)
```

```python
col = ['확진번호', '국적', '정보', '차수', '감염경로']
```

```python
sum1_df = pd.DataFrame(sum1_results, columns=col)
sum1_df
```

```python
sum2 = soup.select('tr.summary.second > td')

sum2_result = []

for td in sum2:
    sum2_txt = td.text
    sum2_result.append(sum2_txt)
    print(sum2_txt)
#     if sum2_txt == '#U17':
#         break
```

```python
sum2_results = []

for group2 in chunker(sum2_result, 5):
    group2 = group2
    sum2_results.append(group2)
    print(group2)
```

```python
col2 = ['blank', '확진날짜', '접촉', '상태', '격리시설']
```

```python
sum2_df = pd.DataFrame(sum2_results, columns=col2)
sum2_df
```

```python
sum1_df = sum1_df.drop(299, axis=0).copy()
sum1_df
```

```python
sum2_df = sum2_df.drop('blank', axis=1).copy()
sum2_df = sum2_df.drop(sum2_df.index[299:]).copy()
sum2_df
```

```python
feature = pd.concat([sum1_df, sum2_df], axis=1)
feature
```
