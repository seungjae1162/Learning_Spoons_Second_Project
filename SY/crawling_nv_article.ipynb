```python
from selenium import webdriver
import time
```

```python
url = "https://news.naver.com/main/ranking/read.nhn?m_view=1&includeAllCount=true&sid1=&rankingType=popular_memo&oid=023&aid=0003510311&date=20200225&type=1&rankingSectionId=100&rankingSeq=1"

driver = webdriver.Chrome('c:/informs/chromedriver.exe')

driver.implicitly_wait(30)
driver.get(url)
```

```python
# 더보기 100번까지만 클릭하기
result = 0
while True:
    try:
        더보기 = driver.find_element_by_css_selector('a.u_cbox_btn_more')
        더보기.click()
        time.sleep(0.000000000000000000000000000000000000000000000000001)
        result = result+1
        if result == 99:
            break 
    except:
        break 
```

```python
html = driver.page_source
```

```python
type(html)
```

```python
print(html[:100])
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
type(soup)
```

```python
# print(soup)
```

```python
comment = soup.select('span.u_cbox_contents')
comment[:3]
```

```python
c = soup.select('span.u_cbox_contents')[0]
c.text
```

```python
result = []
```

```python
for c in comment:
    result.append(c.text)
    print(c.text)
```

```python
result[:3]
```

```python
import pandas as pd
```

```python
result_pd = pd.DataFrame(result)
result_pd.head(3)
```
